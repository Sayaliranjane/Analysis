```python
cols=['Row ID', 'Order ID', 'Ship Date', 'Ship Mode',
       'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State',
       'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category',
       'Product Name', 'Quantity', 'Discount', 'Profit']
store_data.drop(cols,axis=1,inplace=True)

store_data=store_data.sort_values('Order Date')
```

```python
store_data.head()
```

```python
store_data=store_data.groupby('Order Date')['Sales'].sum().reset_index()
store_data['Sales']=round(store_data['Sales'])
store_data.head()
```

```python
store_data['Order Date']=pd.to_datetime(store_data['Order Date'])
store_data.set_index('Order Date',inplace=True)
store_data.index
```

```python
store_data.head()
```

```python
store_data_clean=store_data['Sales'].resample('MS').mean()
store_data_clean['2017']
```

```python
store_data_clean.plot(figsize=(10,5))
plt.show()
```

```python
from pandas import Series
from statsmodels.tsa.stattools import adfuller

result = adfuller(store_data_clean)

print("ADF Statistic: %f" % result[0]) #%f means float

print('P-value: %f' % result[1])

print('Critical Values:')

for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value)) #%.3f means float with 3 decimal point

```

```python
#Decomposing

from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(store_data_clean)
plt.plot(store_data_clean, label = 'Original')
plt.legend(loc = 'best')


trend = decomposition.trend
plt.show()
plt.plot(trend, label = "Trend")
plt.legend(loc = "best")

seasonal = decomposition.seasonal
plt.show()
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc = 'best')

residual = decomposition.resid
plt.show()
plt.plot(residual, label = 'Residual')
plt.legend(loc = 'best')
```

```python
import itertools
p=range(0,2)
d=range(0,2)
q=range(0,2)

pdq=list(itertools.product(p,d,q))
seasonal_pdq=[(X[0],X[1],X[2],12) for X in list(itertools.product(p,d,q))]
print("parameter combination")
print('SARIMAX : {}*{}'.format(pdq[1],seasonal_pdq[1]))
print('SARIMAX : {}*{}'.format(pdq[1],seasonal_pdq[2]))
print('SARIMAX : {}*{}'.format(pdq[2],seasonal_pdq[3]))
print('SARIMAX : {}*{}'.format(pdq[2],seasonal_pdq[4]))
```

```python
import statsmodels.api as sm
```

```python
from pylab import rcParams #Param means Parameter

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(store_data_clean, order=param, 
                                            seasonal_order=param_seasonal, 
                                            enforce_stationarity=False, 
                                            enforce_invertibility= False)
            results = mod.fit()
            print('ARIMA{}x{}12- AIC:{}'.format(param, param_seasonal, results.aic))
        except:
                continue
```

```python
mod = sm.tsa.statespace.SARIMAX(store_data_clean,
                                order=(0, 1, 1),
                                seasonal_order=(0, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
#set forecasts to start at 2017â€“01â€“01 to the end of the data to forecast
pred = results.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = store_data_clean['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast',alpha=1, figsize=(14, 7))
ax.fill_between(pred_ci.index,
pred_ci.iloc[:, 0],
pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
pred_uc = results.get_forecast(steps=13)
pred_ci = pred_uc.conf_int()

ax = store_data_clean.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')

ax.fill_between(pred_ci.index,
pred_ci.iloc[:, 0],
pred_ci.iloc[:, 1], color='k', alpha=.25)

ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')

print(pred_ci)

plt.legend()
plt.show()
```
