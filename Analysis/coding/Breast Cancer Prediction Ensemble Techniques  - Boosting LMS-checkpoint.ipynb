```python
breast_cancer_data.rename(columns={'diagnosis': 'target'}, inplace=True)
```

```python
#diagnosis rename karka target
#Map it to 0 and 1, where Malignant = 1, Benign = 0
breast_cancer_data['target'] = breast_cancer_data['target'] .map({'M': 1, 'B': 0})
```

```python
# adaboost experiments
# create x and y train
X=breast_cancer_data.drop('target',axis=1)
y=breast_cancer_data['target'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.20, 
                                                    random_state = 100)
X_train.head()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# check the average cancer occurence rates in train and test data, should be comparable
print(y_train.mean())
print(y_test.mean())
```

```python
#weak learner
shallow_tree=DecisionTreeClassifier(max_depth=2,random_state=100)
```

```python
shallow_tree.fit(X_train,y_train)

y_pred=shallow_tree.predict(X_test)
score=metrics.accuracy_score(y_test,y_pred)
score
```

```python
param_grid={"base_estimator__max_depth":[2,5],
           "n_estimators":[200,400,600]}
```

```python
tree=DecisionTreeClassifier()
ABC=AdaBoostClassifier(base_estimator=tree,learning_rate=0.6,algorithm="SAMME")
```

```python
folds=3
grid_search_ABC=GridSearchCV(ABC,
                            cv=folds,
                            param_grid=param_grid,
                            scoring='roc_auc',
                            return_train_score=True,
                            verbose=1)
```

```python
grid_search_ABC.fit(X_train,y_train)
```

```python
ABC=grid_search_ABC.best_estimator_
```

```python
ABC.fit(X_train,y_train)
```

```python
predictions_train=ABC.predict(X_train)
```

```python
predictions=ABC.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_train,y_pred=predictions_train))
print(r2_score(y_true=y_test,y_pred=predictions))
```

# XGBoost Algorithm

```python
import xgboost as xgb
from xgboost import XGBClassifier
from xgboost import plot_importance
import gc
%matplotlib inline
```

```python
folds=3
param_grid={'learning_rate':[0.2,0.6],
           'subsample':[0.3,0.6,0.9]}
xgb_model=XGBClassifier(max_depth=2, n_estimators=200)

model_cv=GridSearchCV(estimator=xgb_model,
                     param_grid=param_grid,
                     scoring='roc_auc',
                     cv=folds,
                     verbose=1,
                     return_train_score=True)
```

```python
model_cv.fit(X_train,y_train)
```

```python
cv_results=pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
Xgb=model_cv.best_estimator_
```

```python
Xgb.fit(X_train,y_train)
```

```python
predictions_train=Xgb.predict(X_train)
```

```python
predictions=Xgb.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_train,y_pred=predictions_train))
print(r2_score(y_true=y_test,y_pred=predictions))
```
