# Task-3 Prediction Using Decision Tree ML


## Import Libraries and Packages

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
#load iris dataset
from sklearn.datasets import load_iris
```

```python
iris_tree=load_iris()
```

```python
iris_tree
```

## Check rows and columns

```python
#check total no of rows and cols
iris_tree.data.shape
```

```python
#target variable
print("Classes to predict",iris_tree.target_names)
```

```python
#independent variable
print("Features:",iris_tree.feature_names)
```

```python
#split X and y
X=iris_tree.data
y=iris_tree.target
```

```python
print(X.shape)
print(y.shape)
```

# Model Building

```python
# Importing train-test-split 
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=100)

```

```python
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier
#Fit a model with default DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
#predict y variable
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
#Fit a model with DecisionTreeClassifier criterion as entropy
clf_entropy=DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train,y_train)
```

```python
y_pred_entropy=clf_entropy.predict(X_test)
```

```python
#predict y variable
y_pred_entropy
```

```python
#check accuracy scores of train and test data
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print("Accuracy on train data:",accuracy_score(y_true=y_train,y_pred=clf_entropy.predict(X_train)))
print("Accuracy on test data:",accuracy_score(y_true=y_test,y_pred=y_pred_entropy))
```

```python
#Fit a model with hyperparameters
clf2=DecisionTreeClassifier(criterion='entropy',min_samples_split=50)
clf2.fit(X_train,y_train)
```

```python
#check accuracy scores of train and test data
print("Accuracy on train data:",accuracy_score(y_true=y_train,y_pred=clf2.predict(X_train)))
print("Accuracy on test data:",accuracy_score(y_true=y_test,y_pred=clf2.predict(X_test)))
```
