```python
import pandas as pd
import numpy as np
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")

    
#Import Data
mobile_price_data = pd.read_csv(r"C:\Users\91771\Downloads\MobilePriceRangePrediction\Mobile Price Range Prediction\Dataset\mobile_price.csv")

#Check Data
mobile_price_data.head()
```

```python
mobile_price_data=mobile_price_data[mobile_price_data['px_height']>0]
mobile_price_data=mobile_price_data[mobile_price_data['sc_w']>0]
round(mobile_price_data.describe(),2)
```

```python
for column in mobile_price_data:
    plt.figure()
    mobile_price_data.boxplot([column])
```

```python
from scipy import stats
import numpy as np
z=np.abs(stats.zscore(mobile_price_data['fc']))
threshold=3
print(np.where(z>3))
```

```python
mobile_price_data=mobile_price_data[(z<3)]
mobile_price_data
```

```python
y=mobile_price_data['price_range']
X=mobile_price_data.drop('price_range',axis=1)

import statsmodels.api as sm
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler as mms
scaler=mms()
Columns=list(X_train.columns)
X_train[Columns] = scaler.fit_transform(X_train[Columns])
X_train.head()
```

```python
X_test[Columns]=scaler.transform(X_test[Columns])
X_test
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb=BernoulliNB()

bnb.fit(X_train,y_train)

y_pred_class=bnb.predict(X_test)

y_pred_proba=bnb.predict_proba(X_test)

from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb=MultinomialNB()

nb.fit(X_train,y_train)

y_pred_class=nb.predict(X_test)

y_pred_proba=nb.predict_proba(X_test)

from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
from sklearn.naive_bayes import GaussianNB

gb=GaussianNB()

gb.fit(X_train,y_train)

y_pred_class=gb.predict(X_test)

y_pred_proba=gb.predict_proba(X_test)

from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score

print("Precision score:",precision_score(y_test,y_pred_class,average='weighted'))
print("Recall score:",recall_score(y_test,y_pred_class,average='weighted'))
print("F1 score:",f1_score(y_test,y_pred_class,average='weighted'))
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

from sklearn import metrics

gnb=GaussianNB()

lm=LinearRegression()
lm.fit(X_train,y_train)

for features in np.arange(1,(X_train.shape[1])):
    rfe=RFE(lm,features)
    rfe=rfe.fit(X_train,y_train)
    
    X_train_subset=X_train[list(X_train.columns[rfe.support_])]
    X_test_subset=X_test[list(X_test.columns[rfe.support_])]
    
    gnb.fit(X_train_subset,y_train)
    
    y_pred_class=gnb.predict(X_test_subset)
    
    y_pred_proba=gnb.predict_proba(X_test_subset)
    
    print("for",features,"features being selected the accuracy is",metrics.accuracy_score(y_test,y_pred_class))
```

```python
rfe=RFE(lm,10)
rfe=rfe.fit(X_train,y_train)
    
X_train_subset=X_train[list(X_train.columns[rfe.support_])]
X_test_subset=X_test[list(X_test.columns[rfe.support_])]
    
```

```python
from sklearn.model_selection import GridSearchCV

gnb=GaussianNB()

param_grid={'var_smoothing':[0.000000001,0.00000001,0.0000001,0.000001,0.00001,0.0001]}

nb_gscv=GridSearchCV(gnb,param_grid,
                    scoring="accuracy",
                    cv=7,
                    n_jobs=-1,
                    verbose=20)
nb_gscv.fit(X_train_subset,y_train)
```

```python
best_model=nb_gscv.best_estimator_
best_model
```

```python
nb_gscv.best_score_
```

```python
from sklearn.metrics import accuracy_score

y_pred=best_model.predict(X_test_subset)
print(accuracy_score(y_test,y_pred))
```
