```python
cust_data_clean=cust_data.dropna()
cust_data_clean.shape

y=cust_data_clean['Gender']
X=cust_data_clean.drop('Gender',axis=1)

from sklearn.preprocessing import MinMaxScaler as mms

scaler=mms()
Columns=list(X.columns)
X[Columns]=scaler.fit_transform(X[Columns])
X
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
knn2=KNeighborsClassifier()

param_grid={'n_neighbors':[49,50,51]}

knn_gscv=GridSearchCV(knn2,param_grid,
                     scoring="accuracy",
                     cv=7,
                     n_jobs=-1,
                     verbose=20)

knn_gscv.fit(X,y)
```

```python
best_model=knn_gscv.best_estimator_
```

```python
best_model
```

```python
knn_gscv.best_score_
```

```python
from sklearn.metrics import accuracy_score

y_pred=best_model.predict(X)
print(accuracy_score(y,y_pred))
```

```python
Na_cust_data=cust_data[cust_data['Gender'].isnull()]
Na_cust_data.shape
```

```python
Na_cust_data=Na_cust_data.drop('Gender',axis=1)

Columns=list(Na_cust_data.columns)
Na_cust_data[Columns]=scaler.transform(Na_cust_data[Columns])
Na_cust_data.head()
```

```python
cust_data['Gender']=np.where(cust_data['Gender']=='F',0,1)
cust_data['Gender'].describe()
```

```python
#Modelling

kmeans=KMeans(n_clusters=4, max_iter=50)
kmeans.fit(cust_data)
```

```python
kmeans.labels_
```

```python
#ssd=sumofsquared_deviance
#elbow method/curve_fit
ssd=[]
range_n_clusters=[2,3,4,5,6,7,8,10,12]
for num_clusters in range_n_clusters:
    kmeans=KMeans(n_clusters=num_clusters,max_iter=50)
    kmeans.fit(cust_data)
    
    ssd.append(kmeans.inertia_)
plt.plot(ssd)
```

```python
range_n_clusters=[2,3,4,5,6,7,8,10,12]
for num_clusters in range_n_clusters:
    kmeans=KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(cust_data)
        
    cluster_labels=kmeans.labels_
        
    silhouette_avg=silhouette_score(cust_data,cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters,silhouette_avg))
```

```python

kmeans=KMeans(n_clusters=3, max_iter=50)
kmeans.fit(cust_data)
```

```python
cust_data['cluster_id']=kmeans.labels_
cust_data.head()
```

```python
gby=cust_data.groupby(['cluster_id'])
gby.first()
```

```python
fig,ax=plt.subplots(figsize=(8,8))
sns.heatmap(gby,cmap="coolwarm")
```
