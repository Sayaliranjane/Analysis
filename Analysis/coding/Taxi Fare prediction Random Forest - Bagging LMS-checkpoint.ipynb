```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

taxi_data=pd.read_csv(r"C:\Users\91771\Downloads\TaxiFarePrediction\Taxi Fare Prediction\Dataset\TaxiFare.csv")
taxi_data.head()

```

```python
#Obs :
#1. Ammount cannot be negative we need to remove that data
#2. Passengers cannot be 0
#3. The numbers range from -90 to 90 for latitude and -180 to 180 for longitude.
taxi_data.describe()
```

```python
#1 Ammount cannot be negative we need to remove that data also minimum fare for NY cab is $ 2.5
# so any value below that is disqualified



taxi_data=taxi_data.drop(taxi_data[taxi_data['amount']<2.5].index,axis=0)
taxi_data.describe()
```

```python
#2. Passengers cannot be 0
taxi_data=taxi_data[taxi_data.no_of_passenger>0]
taxi_data.describe()
```

```python
#3. The numbers range from -90 to 90 for latitude and -180 to 180 for longitude.
taxi_data=taxi_data[(taxi_data.latitude_of_pickup<180)]
taxi_data.describe()
```

```python
taxi_data.shape
```

```python
#Creating a new feature called Distance
def Dist_cal(x2,x1,y2,y1):
    result=((((x2-x1)**2)+((y2-y1)**2))**0.5)
    return result
```

```python
taxi_data['distance']=abs(Dist_cal(taxi_data.longitude_of_pickup,taxi_data.latitude_of_pickup,taxi_data.longitude_of_dropoff,taxi_data.latitude_of_dropoff)-abs((Dist_cal(taxi_data.longitude_of_pickup,taxi_data.latitude_of_pickup,taxi_data.longitude_of_dropoff,taxi_data.latitude_of_dropoff)).mean()))
taxi_data['distance'].describe()
```

```python
taxi_data['distance'].hist()
```

```python
taxi_data['amount'].hist()
```

```python
#Most certainly there seems to be some outlier in the distances lets remove them
taxi_data=taxi_data[taxi_data['distance']<20]
taxi_data.describe()
```

```python
plt.scatter(taxi_data['amount'],taxi_data['distance'])
```

```python
taxi_data=taxi_data[taxi_data['distance']<8]
taxi_data.describe()
```

```python
taxi_data.info()
```

```python
taxi_data.date_time_of_pickup=pd.to_datetime(taxi_data.date_time_of_pickup,utc=True,infer_datetime_format=True)
```

```python
taxi_data.info()
```

```python
taxi_data['Hour']=taxi_data['date_time_of_pickup'].apply(lambda time:time.hour)
taxi_data['Month']=taxi_data['date_time_of_pickup'].apply(lambda time:time.month)
taxi_data['Day of Week']=taxi_data['date_time_of_pickup'].apply(lambda time:time.dayofweek)
taxi_data['Year']=taxi_data['date_time_of_pickup'].apply(lambda time:time.year)
taxi_data['Date']=taxi_data['date_time_of_pickup'].apply(lambda time:time.day)

del taxi_data['date_time_of_pickup']
taxi_data.head()
```

```python
del taxi_data['unique_id']

taxi_data.head()
```

```python
from sklearn.model_selection import train_test_split as tts

X=taxi_data.drop('amount',axis=1)
y=taxi_data['amount']

X_train,X_test,y_train,y_test=tts(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
predictions_train=rfr.predict(X_train)
```

```python
predictions=rfr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_train,y_pred=predictions_train))
print(r2_score(y_true=y_test,y_pred=predictions))
```

```python

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor()

params={'n_estimators':[100,200,300,500,700,1000],
       'max_features':[6,8,10,12],
       'bootstrap':[True,False],
       'max_depth':[None,5,10,15,20,30,50,70],
       'min_samples_leaf':[1,2,3,4],
       'min_samples_split':[13,14,15,16,17]}

random_search=RandomizedSearchCV(reg,
                                cv=7,
                                param_distributions=params,
                                scoring='neg_mean_absolute_error',
                                n_iter=10,
                                n_jobs=-1,
                                verbose=20)
random_search.fit(X_train,y_train)
```

```python
rf=random_search.best_estimator_
```

```python
rf.fit(X_train,y_train)
```

```python
predictions_train=rf.predict(X_train)
```

```python
predictions=rf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_train,y_pred=predictions_train))
print(r2_score(y_true=y_test,y_pred=predictions))
```
