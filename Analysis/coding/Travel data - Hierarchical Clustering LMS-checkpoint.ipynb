```python
google_data_clean=google_data.dropna()
```

```python
google_data_clean['local_services']=google_data_clean['local_services'].astype('float64')
google_data_clean.info()
```

```python
# single linkage
mergings=linkage(google_data_clean,method="single",metric="euclidean")
dendrogram(mergings)
plt.show()
```

```python
# complete linkage
mergings=linkage(google_data_clean,method="complete",metric="euclidean")
dendrogram(mergings)
plt.show()
```

```python
cluster_labels=cut_tree(mergings,n_clusters=5).reshape(-1, )
cluster_labels
```

```python
google_data_clean['cluster_labels']=cluster_labels
google_data_clean.head()
```

```python
gby=google_data_clean.groupby(['cluster_labels'])
gby.first()
```

```python
gby=pd.DataFrame(gby.first())
gby=gby.T
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(gby,cmap="coolwarm")
```
